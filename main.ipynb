```python
# 쿠키 정보 가져오기 & 헤더 선언
import time
time.sleep(2)   # 쿠키가 생성될 동안 대기

cookies = None
try:
    cookies = driver.get_cookies()
    print("쿠키 정보 가져옴")
except:
    print("[ERROR!] 쿠키 가져오기 실패! 같은 오류가 반복되면 코드를 확인해주세요.")
    stopChrome()

cookies_json = {}

for obj in cookies:
    name = obj['name']
    value = obj['value']
    cookies_json[name] = value


HEADERS = {
    "Cookie": ""\
        "userId="+ cookies_json['userId'] +";"\
        "Token="+ cookies_json['Token'] +";"\
        "iuid="+ cookies_json['iuid'] +";"\
        "name="+ cookies_json['name'] +";"\
        "nickname="+ cookies_json['nickname'] +";"\
        "memberGradeCd="+ cookies_json['memberGradeCd'] +";"\
        "JSESSIONID="+ cookies_json['JSESSIONID'] +";"\
        "_ga="+ cookies_json['_ga'] +";"\
        "_ga_E5HXF8M05M="+ cookies_json['_ga_E5HXF8M05M'] +";"\
        "_gid="+ cookies_json['_gid'] +";"\
        "__gads="+ cookies_json['__gads'] +";"\
        "__gpi="+ cookies_json['__gpi'] +";"\
}
```

```python
# 위문편지함 HTML 데이터 가져오기
import requests
from bs4 import BeautifulSoup

home_req = requests.get(HOME_URL, headers=HEADERS)
home_html = None

if home_req.status_code == 200:
    home_raw_html = home_req.text
    home_soup = BeautifulSoup(home_raw_html, 'html.parser')
    home_html = home_soup.find_all('a', class_='receive-card-box')
else:
    print(f"[ERROR!] 웹페이지 접속 실패!(code: {home_req.status_code}) 같은 오류가 반복되면 코드를 확인해주세요.")
    stopChrome()
```

```python
# 사용자 정보 추출

# 내 고유값 & 나에게 편지를 보낸 사용자의 고유값 추출
my_uid = home_html[0].attrs['href'].split('\'')[1]

uid_list = []
for html_a in home_html:
    href = html_a.attrs['href']
    uid = href.split(',')[1].split('\'')[1]
    uid_list.append(uid)

# 사용자 ID 추출
user_list = {}
user_data_json = None

for uid in uid_list:
    REQ_DATA = {
        "traineeMgrSeq": my_uid,
        "curPage": '1',
        "sendIuid": uid,
    }
    user_data_req = requests.post(LETTER_USER_URL, headers=HEADERS, data=REQ_DATA)

    if user_data_req.status_code != 200:
        print(f"[ERROR!] 사용자 정보 가져오기 실패!(code: {user_data_req.status_code}) 같은 오류가 반복되면 코드를 확인해주세요.")
        stopChrome()

    user_data_json = user_data_req.json()

    name = user_data_json['listResult'][0]['cafeNickname']  # 사용자 이름
    total = user_data_json['iListTotCnt']                   # 사용자가 보낸 전체 편지 수
    page = user_data_json['iTotPageCnt']                    # 위문편지함 전체 페이지 수 (리스트 추출 시 사용)

    user_list_item = {
        'name': name,
        'relation': 'undefined',
        'total': total,
        'page': page,
        'letterList': None
    }

    user_list[uid] = user_list_item

# print(user_list)
```

```python
# 편지 리스트 추출
for uid in uid_list:
    letter_list = []
    page_cnt  = 1

    while page_cnt <= user_list[uid]['page']:
        REQ_DATA = {
            "traineeMgrSeq": my_uid,
            "curPage": page_cnt,
            "sendIuid": uid,
        }
        user_data_req = requests.post(LETTER_USER_URL, headers=HEADERS, data=REQ_DATA)
        page_cnt += 1

        if user_data_req.status_code != 200:
            print(f"[ERROR!] 편지 리스트 가져오기 실패!(code: {user_data_req.status_code}) 같은 오류가 반복되면 코드를 확인해주세요.")
            stopChrome()

        user_data_json = user_data_req.json()

        for listResult in user_data_json['listResult']:
            letterId = listResult['sympathyLetterMgrSeq']
            date = listResult['uptDate']
            time = listResult['uptTime']

            letter_list_item = {
                'letterId': str(letterId),
                'date': date,
                'time': time,
            }
            letter_list.append(letter_list_item)
    
    user_list[uid]['letterList'] = letter_list

# print(user_list)
```

```python
# 실제 편지 내용 추출
letter_data = {}

# CDN 업로드 파일 이미지 URL 생성
def fn_getFileSrcUrl(obj):
    serverMode = "real"
    if serverMode == "real":  cdnServerMode = "p"
    elif serverMode == "stg": cdnServerMode = "s"

    sFileType = obj['fileTypeCd']
    sImageThumbnailSize = ""

    if obj['filePath'] == "": return ""

    sFilePath = sCdnFileSrcUrl + obj['filePath'] + obj['savedFileNm']

    if (obj['extNm'] is not None) and (obj['extNm'] != "") and type(obj['extNm']) != "undefined":
        sFilePath += "." + obj['extNm']

    if (sFileType == sImageCommCd):
        if (obj['thumb'] is None) or (obj['thumb'] == "") or type(obj['thumb']) == "undefined":
            sImageThumbnailSize = "0x0"
        else:
            sImageThumbnailSize = obj['thumb']

        sFilePath += "?d=" + sImageThumbnailSize + "&s=" + cdnServerMode
    else:
        sFilePath += "?s=" + cdnServerMode

    return sFilePath

for uid in user_list:
    for letter_item in user_list[uid]['letterList']:
        letterId = letter_item['letterId']
        REQ_DATA = {
            'curPage': 1,
            'sympathyLetterMgrSeq': letterId,
            'letterType': 'receive'
        }
        letter_req = requests.post(LETTER_URL, headers=HEADERS, data=REQ_DATA)
        letter_received_html = None
        letter_img_path = ''

        if letter_req.status_code == 200:
            letter_req_html = letter_req.text   # string
            letter_req_soup = BeautifulSoup(letter_req_html, 'html.parser')
            letter_received_tbody = letter_req_soup.find('tbody')
            letter_received_td = letter_received_tbody.find_all('td')

            relation = letter_received_td[1].get_text().split('(')[1].split(')')[0]
            letter_title = letter_received_td[2].get_text().replace('접수완료', '').replace('출력완료', '').strip()
            letter_content_html = str(letter_received_td[3].find('div')).replace('<div class="edit-cont">', '<div>').replace('\xa0', '').replace('\n', '')
            letter_content_text = letter_received_td[3].find('div').get_text().replace('\xa0', '').replace('\n', '')

            letter_img = ''
            # img태그가 있다면 제거
            if '<img' in letter_content_html:
                letter_content_html = letter_content_html.split('<img')[0] + '</div>'     
                
                letter_img_data_html = letter_req_html.split("function fn_getImgUrl() {")[1].split('}')[0].split(':')[1:5]
                letter_img_data = {
                    'fileTypeCd': letter_img_data_html[0].split('\'')[1],
                    'filePath': letter_img_data_html[1].split('\'')[1],
                    'savedFileNm': letter_img_data_html[2].split('\'')[1],
                    'extNm': letter_img_data_html[3].split('\'')[1],
                    'thumb': '1000x0',
                }
                letter_img_url = fn_getFileSrcUrl(letter_img_data)
                letter_img_response = requests.get(letter_img_url)
                letter_img_path = './img/' + uid + letterId + '.png'
                letter_img = open(letter_img_path, 'wb')
                letter_img.write(letter_img_response.content)
                letter_img.close()
        else:
            print(f"[ERROR!] 편지 내용 가져오기 실패!(code: {letter_req.status_code}) 같은 오류가 반복되면 코드를 확인해주세요.")
            stopChrome()
        
        user_list[uid]['relation'] = relation

        letter_data_item = {
            'uid': uid,
            'letterTitle': letter_title,
            'letterContentHtml': letter_content_html,
            'letterContentText': letter_content_text,
            'imgPath': letter_img_path,
        }

        letter_data[letterId] = letter_data_item

# print(user_list)
# print(letter_data)
```

```python
# html 파일로 저장
output_html = ''
output_html += HTML_HEAD

user_list_html = user_list.copy()
letter_data_html = letter_data.copy()
output_html += 'user_list=' + str(user_list_html) + ';'
output_html += 'letter_data=' + str(letter_data_html) + ';'

output_html += HTML_BODY

output_html_file = open('output.html', 'w', encoding='utf-8-sig')
output_html_file.write(output_html)
output_html_file.close()
```
